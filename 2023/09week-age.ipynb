```python
# 성별 인구 구조
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
male=[]
female=[]

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            male.append(int(i))
        for i in row[106:207]:
            female.append(int(i))

f.close()
plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')
plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
male=[]
female=[]

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:207]:
            female.append(int(i))

f.close()
plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')
plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 제주도 여성 비율이 더 높을까

import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
male=[]
female=[]

name = input('찾고 싶은 지역의 이름을 입력하세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            i = i.replace(',', "")
            male.append(-int(i))
        for i in row[106:207]:
            i = i.replace(',', "")
            female.append(int(i))
        break

f.close()
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')
plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#제주특별자치도 성별 인구 합계 구하기


# 제주도 여성 비율이 더 높을까

import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('찾고 싶은 지역의 이름을 입력하세요 : ')

for row in data:
    if name in row[0]:
        male = 0
        female = 0
        for i in row[3:104]:
            i = i.replace(',', "")
            male = male + int(i)
        for i in row[106:207]:
            i = i.replace(',', "")
            female = female + int(i)
        break

f.close()

print('남성 인구 합: ', male, '명')
print('여성 인구 합: ', female, '명')
```
